```python
from openmm import unit
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# RampUpdateReporter

start_value = 5.0 #* unit.nanometer
end_value = 6.0 #* unit.nanometer
update_interval = 10
steps =  100

current_value = start_value + (end_value - start_value) * (update_interval/steps) 
vals=[]

for current_step in range(update_interval, steps, update_interval):
    
    current_value += (end_value - start_value) * (update_interval/steps) 
    vals.append(current_value)

    print(current_step, current_value)
```

```python
# TriangleUpdateReporter

start_value = 5.0 #* unit.nanometer
end_value = 6.0 #* unit.nanometer
update_interval = 10
switch_step = 30 #switch step must be greater than update interval
steps =  100

current_value = start_value
vals =[]

for current_step in range(0, steps, update_interval):
    
    if current_step < switch_step:
    
        current_value += (end_value - start_value) * (update_interval/switch_step) 
        
    else:
        current_value -= (end_value - start_value) * (update_interval/(steps - switch_step)) 
        
    vals.append(current_value)

    print(current_step, current_value)
```

```python
# SinusoidalSquaredUpdateReporter

amplitude = 1
oscillation_steps = 20
period =  np.pi/oscillation_steps
phase_shift = np.pi/2
vertical_shift = 0
update_interval = 1
steps = 100
start_value = 5.0

vals = []

current_value = start_value

for current_step in range(0, steps, update_interval):
    
    
    current_value = amplitude * start_value * math.sin(period*current_step + phase_shift)**2 + vertical_shift

    
    vals.append(current_value)

    print(current_step, current_value)
```

```python
plt.plot(vals)
plt.show()
```
